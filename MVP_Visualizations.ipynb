```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df18 = pd.read_csv("2018-DODR.csv")
df18
```

```python
df19 = pd.read_csv("2019-DODR.csv")
df19
```

```python
dfmap = pd.read_csv("2018-2019-DODR-MAP.csv")
dfmap
```

```python
df19 = df19.dropna()
df19
```

```python
dfmap = dfmap.dropna()
dfmap
```

### Bar chart of overdose deaths in 2018

```python
deathsbystate18 = df18[['Location','2018 Number of Deaths']]
deathsbystate18
```

```python
df18.dtypes
```

```python
df18.plot.bar(x='Location', y='2018 Number of Deaths', legend=False,figsize=(15,10))
plt.ylabel('Number of Deaths')
plt.title('Number of drug overdose deaths per state in 2018')
```

### Bar chart of overdose deaths in 2019

```python
df19.plot.bar(x='Location', y='2019 Number of Deaths', legend=False,figsize=(15,10))
plt.ylabel('Number of Deaths')
plt.title('Number of drug overdose deaths per state in 2019')
```

### Scatter plot of changes in overdose deaths (increasing/decreasing/stable) per state from 2018 to 2019 

```python
dfmap.plot.scatter(x='Location', y='Statistically Significant Change ', figsize=(15,10))
plt.title('Changes in drug overdose deaths from 2018 to 2019 per state')
plt.xticks(rotation=90)
plt.show()
```

### Bar chart showing top 5 states for increases in overdose deaths 

```python
top5inc = dfmap[dfmap['Statistically Significant Change '] == 'Increase'].sort_values(by=['Percent Change from 2018-2019'], ascending=False).head(5)
top5inc
```

```python
top5inc.plot.bar(x='Location', y='Percent Change from 2018-2019', legend=False, figsize=(15,10))
plt.title('Top five states for increasing drug overdose deaths from 2018-2019')
plt.ylabel('Percent increase')
plt.xlabel('State')
plt.xticks(rotation=45)
```

### Bar chart showing top 5 states for a decrease in overdose deaths 

```python
top5dec = dfmap[dfmap['Statistically Significant Change '] == 'Decrease'].sort_values(by=['Percent Change from 2018-2019'], ascending=False).head(5)
top5dec
```

```python
top5dec.plot.bar(x='Location', y='Percent Change from 2018-2019', legend=False, figsize=(15,10))
plt.title('Decrease in drug overdose deaths per state from 2018-2019')
plt.ylabel('Percent change')
plt.xticks(rotation=360)
plt.xlabel('State')
```

### Bar chart showing changes from 2018 to 2019 in overdose deaths in Arizona

```python
Az18 = df18[df18['Location'] == 'Arizona']['2018 Number of Deaths']
Az18
```

```python
Az19 = df19[df19['Location'] == 'Arizona']['2019 Number of Deaths']
Az19
```

```python
dfmerge = pd.merge(df18,df19)
dfmerge
```

```python
arizona = dfmerge[dfmerge['Location'] == 'Arizona'][['2018 Number of Deaths','2019 Number of Deaths']]

arizona
```

```python
arizona.plot.bar(figsize=(15,10))
plt.ylabel('Number of Deaths')
plt.xlabel('Year')
plt.title('Number of drug overdose deaths in Arizona from 2018-2019')
```
